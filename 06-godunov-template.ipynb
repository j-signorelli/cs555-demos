```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def u0(x):
    u = np.zeros_like(x)
    u[np.where((x>=0.25) & (x<=0.5))[0]] = 1.0
    return u
xx = np.linspace(0, 2, 1000)
plt.plot(xx, u0(xx))
```

```python
n=4
# n = 4
# h = 1/4
#             0       1       2       3
#     0   |   1       2       3       4   |   5
# |-------|-------|-------|-------|-------|-------|
#         a      a+h     a+2h    a+3h     b
#    a+h/2   a+3h/2  a+5h/2  a+7h/2
# periodic: u[a]=u[b]
#        +: u[0]=u[4]
#        +: u[5]=u[1]
a=0
b=2
h=(b-a)/n
print(h)
assert np.abs(h-0.5)<1e-14
x = np.arange(a+h/2, b, h) # middle of the cells
u = np.zeros(n+2)
u[1:-1] = u0(x)  # initial cell averages
u[0] = u[1]
u[-1] = u[-2]
```
